#### 7. Mengecek outlier data

```python
# Mengecek visualisasi dengan boxplot untuk nilai outlier data
num_cols = [
    'Total Land Area (acres)', 'Number of Cows', 'Product ID', 'Quantity (liters/kg)', 'Price per Unit', 'Total Value', 
    'Shelf Life (days)', 'Quantity Sold (liters/kg)', 'Price per Unit (sold)', 'Approx. Total Revenue(INR)', 'Quantity in Stock (liters/kg)', 
    'Minimum Stock Threshold (liters/kg)', 'Reorder Quantity (liters/kg)'
]

# 1. Tentukan mau berapa kolom ke samping
n_cols = 5
n_rows = math.ceil(len(num_cols) / n_cols) # Menghitung otomatis butuh berapa baris ke bawah

# 2. Buat kanvas besar (figure) dengan ukuran grid baris x kolom yang ditentukan
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(15, n_rows * 3.5))

# 3. Ratakan koordinat axes (dari matriks 2D menjadi 1D) biar gampang di-looping
axes = axes.flatten()

# 4. Loop untuk menggambar boxplot di setiap slot
for i, col in enumerate(num_cols):
    sns.boxplot(x=df[col], ax=axes[i]) # Perhatikan ada tambahan ax=axes[i]
    axes[i].set_title(col, fontsize=10)
    axes[i].set_xlabel('') # Opsional: mengosongkan label x biar gak tumpang tindih

# 5. Karena 5x3 = 15 slot sedangkan kolom kita cuma 13, hapus 2 slot kosong sisanya
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# 6. Atur layout agar rapi dan tidak tumpang tindih, lalu tampilkan
plt.tight_layout()
plt.show()
```

```python
# Menghitung total seluruh baris untuk nilai outlier dan jumlah outlier serta persentasenya
# 1. Hitung total baris data
total_data = df.shape[0]

# 2. Daftar kolom numerik yang dicek outlier-nya
kolom_numerik = [
   'Total Land Area (acres)', 'Number of Cows', 'Product ID', 'Quantity (liters/kg)', 'Price per Unit', 'Total Value', 
    'Shelf Life (days)', 'Quantity Sold (liters/kg)', 'Price per Unit (sold)', 'Approx. Total Revenue(INR)', 'Quantity in Stock (liters/kg)', 
    'Minimum Stock Threshold (liters/kg)', 'Reorder Quantity (liters/kg)'
]

# 3. Lakukan perhitungan dengan looping dan simpan hasilnya ke list
outlier_summary = []

for col in kolom_numerik:
    # Hitung Q1, Q3, dan IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Hitung batas bawah dan batas atas
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    
    # Hitung jumlah dan persentase outlier
    jumlah_outlier = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    pct_outlier = (jumlah_outlier / total_data) * 100
    
    # Masukkan data ke dalam list format dictionary
    outlier_summary.append({
        'Nama Kolom': col,
        'Batas Bawah': round(lower_bound, 2),
        'Batas Atas': round(upper_bound, 2),
        'Jumlah Outlier (Baris)': jumlah_outlier,
        'Persentase Outlier (%)': f"{pct_outlier:.2f}%"
    })

# 4. Ubah list menjadi DataFrame agar berbentuk tabel
df_outlier = pd.DataFrame(outlier_summary)

# 5. Tampilkan tabelnya
df_outlier
```

Interpretasi :
- Dari 13 kolom numerik diketahui ada 5 kolom yang terdapat nilai outlier, yaitu total value, Shelf Life (days), Quantity Sold (liters/kg), Approx. Total Revenue(INR), Quantity in Stock (liters/kg)
- Untuk kolom tersebut wajar jika adanya nilai yang berbeda jauh sehingga tidak perlu diubah

#### Kelompok 3 : Product Performance Analysis

#### 1. Sell Through Rate

```python
# Mengukur seberapa besar stok yang berhasil terjual
df_clean['Sell_Through_Rate'] = (
    df_clean['Quantity Sold (liters/kg)']
    /
    df_clean['Quantity (liters/kg)']
)
df_clean[['Sell_Through_Rate']].head()
```

```python
# statistika deskriptif dari sell through rate
round(df_clean['Sell_Through_Rate'].describe(), 2)
```

```python
# Mengecek jumlah data yang terjual melebih 100%
(df_clean['Sell_Through_Rate'] > 1).sum()
```

```python
# Menghitung persentase produk yang terjual melebihi 100%
round(
    (df_clean['Sell_Through_Rate'] > 1).mean() * 100,
    2
)
```

Keterangan : sell_through_rate untuk mengukur seberapa besar stok yang berhasil terjual. Dengan nilai yang didapatkan dikalikan dengan 100%, sebagai berikut :
- Jika nilainya 1 mengartikan 100% stok terjual (semua stok terjual)
- Jadi 0.088 maka barang yang terjual 8,8% dari stok begitupun selanjutnya
- Ada 12.71% produk yang penjualannya melebih stok yang ada, artinya perusahaan perlu evaluasi untuk stok barangnya yang jumlah barang habis terjual melebihi jumlah barang yang ada di stok. Sehingga perusahaan perlu menambah stok

#### 2. Product Movement

```python
# Membuat klasifikasi sesuai nilai sell through rate
def movement(x):
    if x> 1:
        return 'Over Moving'
    elif x >= 0.8:
        return 'Fast Moving'
    elif x >= 0.5:
        return 'Medium Moving'
    else:
        return 'Slow Moving'

df_clean['Product_Movement'] = (
    df_clean['Sell_Through_Rate']
    .apply(movement)
)
df_clean[['Product_Movement']].head()
```

```python
# Menampilkan jumlah data masing-masing klasifikasi produk movement
print(df_clean['Product_Movement'].value_counts())
```

```python
# Menghitung nilai persentase perhitungan
round(df_clean['Product_Movement'].value_counts(normalize=True) * 100, 2)
```

Interpretasi :
- Dataset tersebut diketahui untuk performa produknya mayoritas produk yang terjual itu dibawah 50% barang yang terjual dengan persentase jumlahnya 54,7% produk slow moving dan untuk produk yang dapat terjual banyak hanya 9,81% (fast moving) dengan kriterian persentase barang yang terjual diatas 80% sedangkan ada 12,71% barang yang dijual melebih stok yang disediakan.

#### Kelompok 4 : Inventory Analysis

#### 1. Inventory Status

```python
# Menghitung nilai status inventory produk
df_clean['Inventory_Status'] = np.where(
    df_clean['Quantity in Stock (liters/kg)']
    < df_clean['Minimum Stock Threshold (liters/kg)'],
    'Restock Needed',
    'Safe'
)
df_clean[['Inventory_Status']].head()
```

```python
# Melihat jumlah data yang yang masih aman dan butuh stok tambahan
print(df_clean['Inventory_Status'].value_counts())
```

```python
# Menghitung nilai persentasenya
round(df_clean['Inventory_Status'].value_counts(normalize=True) * 100, 2)
```

Interpretasi :
- Kolom ini digunakan untuk mengetahui kebutuhan untuk menambah stok produk atau restock analysis
- Sebanyak 78,84% untuk stock produk sudah aman dan 21,16% diperlukan untuk menambah stok product

#### 2. Stock Gap

```python
# Melihat stok barang yang aman atau kurang aman (positif=stok aman, negtif = stok kurang
df_clean['Stock_Gap'] = (
    df_clean['Quantity in Stock (liters/kg)']
    -
    df_clean['Minimum Stock Threshold (liters/kg)']
)
df_clean[['Stock_Gap']].head()
```

```python
# Mengecek nilai statistikanya
round(df_clean['Stock_Gap'].describe(), 2)
```

Interpretasi :
- Jika nilai stock gapnya positif maka menyatakan bahwa stok yang tersedia aman
- Sedangkan jika nilai stock gapnya negatif maka menyatakan bahwa terjadi kekurangan stock
- Positif ➜ stok masih di atas batas minimum
- Nol ➜ stok tepat di batas minimum
- Negatif ➜ stok sudah di bawah batas minimum (perlu restock)

#### 3. Remaining Shelf Life

```python
# Mengecek expired  produk pada saat dilakukan pencatatan data
df_clean['Remaining_Shelf_Life'] = (
    df_clean['Expiration Date']
    -
    df_clean['Date']
).dt.days
df_clean[['Remaining_Shelf_Life']].head()
```

#### 4. Expiry_Risk

```python
# Membuat klasifikasi untuk jangka expired produk
def expiry_risk(x):
    if x < 0:
        return 'Expired'
    elif x <= 7:
        return 'High Risk'
    elif x <= 30:
        return 'Medium Risk'
    else:
        return 'Low Risk'

df_clean['Expiry_Risk'] = (
    df_clean['Remaining_Shelf_Life']
    .apply(expiry_risk)
)
```

```python
# Mengecek jumlah data sesuai klasifikasi janka expirednya
print(df_clean['Expiry_Risk'].value_counts())
```

Interpretasi :
Secara operasional, stok produk yang masih layak jual sebenarnya didominasi oleh kategori aman (Low Risk). Namun secara kualitas data, dataset ini memiliki masalah serius (data anomaly) karena lebih dari 50% data masuk dalam kategori **Expired**. Ini perlu evaluasi dari perusahan supaya mencatat produk lebih awal dan tidak mendekati atau bahkan lewat dari tanggal kadaluwarsanya.
